In the following, we will perform some data cleaning on the raw text of the tweets.<br>

**Then we define custom functions to clean the text of the tweets.**

This code block defines several custom functions to clean tweets. The strip_emoji() function uses the demoji library to remove emojis from the text. The strip_all_entities() function removes punctuations, links, mentions, and new line characters. The clean_hashtags() function removes hashtags from the end of the sentence and keeps those in the middle of the sentence. The filter_chars() function filters special characters such as $ and & present in some words. The remove_mult_spaces() function removes multiple spaces from the text.





```python
##CUSTOM DEFINED FUNCTIONS TO CLEAN THE TWEETS

#Clean emojis from text
def strip_emoji(text):
    return demoji.replace(text, '') #remove emoji

#Remove punctuations, links, mentions and \r\n new line characters
def strip_all_entities(text): 
    text = text.replace('\r', '').replace('\n', ' ').replace('\n', ' ').lower() #remove \n and \r and lowercase
    text = re.sub(r"(?:\@|https?\://)\S+", "", text) #remove links and mentions
    text = re.sub(r'[^\x00-\x7f]',r'', text) #remove non utf8/ascii characters such as '\x9a\x91\x97\x9a\x97'
    banned_list= string.punctuation + 'Ã'+'±'+'ã'+'¼'+'â'+'»'+'§'
    table = str.maketrans('', '', banned_list)
    text = text.translate(table)
    return text

#clean hashtags at the end of the sentence, and keep those in the middle of the sentence by removing just the # symbol
def clean_hashtags(tweet):
    new_tweet = " ".join(word.strip() for word in re.split('#(?!(?:hashtag)\b)[\w-]+(?=(?:\s+#[\w-]+)*\s*$)', tweet)) #remove last hashtags
    new_tweet2 = " ".join(word.strip() for word in re.split('#|_', new_tweet)) #remove hashtags symbol from words in the middle of the sentence
    return new_tweet2

#Filter special characters such as & and $ present in some words
def filter_chars(a):
    sent = []
    for word in a.split(' '):
        if ('$' in word) | ('&' in word):
            sent.append('')
        else:
            sent.append(word)
    return ' '.join(sent)

def remove_mult_spaces(text): # remove multiple spaces
    return re.sub("\s\s+" , " ", text)
```

This code block cleans the tweets in the OriginalTweet column of a pandas dataframe df using the custom functions defined earlier (remove_mult_spaces(), filter_chars(), clean_hashtags(), strip_all_entities(), strip_emoji()). The cleaned tweets are then stored in a new list texts_new.





```python
texts_new = []
for t in df.OriginalTweet:
    texts_new.append(remove_mult_spaces(filter_chars(clean_hashtags(strip_all_entities(strip_emoji(t))))))
```

This code block is preprocessing textual data for a machine learning model. It applies several functions to clean up the text, such as removing emojis, punctuation, links, mentions, special characters, and extra spaces. It then stores the cleaned text in a new list for the training and testing data separately.





```python
texts_new_test = []
for t in df_test.OriginalTweet:
    texts_new_test.append(remove_mult_spaces(filter_chars(clean_hashtags(strip_all_entities(strip_emoji(t))))))
```

# BERT Sentiment Analysis

We already performed a basic analyis of the tokenized sentences, now we just need to define a custom tokenizer function and call the encode_plus method of the BERT tokenizer.

```python
MAX_LEN=128
```

This code defines a function tokenize that takes in a list of textual data and uses the tokenizer object to encode the data as input IDs and attention masks. The MAX_LEN variable is set as a parameter for the maximum length of input sequences. The function returns two NumPy arrays: input_ids and attention_masks.





```python
def tokenize(data,max_len=MAX_LEN) :
    input_ids = []
    attention_masks = []
    for i in range(len(data)):
        encoded = tokenizer.encode_plus(
            data[i],
            add_special_tokens=True,
            max_length=MAX_LEN,
            padding='max_length',
            return_attention_mask=True
        )
        input_ids.append(encoded['input_ids'])
        attention_masks.append(encoded['attention_mask'])
    return np.array(input_ids),np.array(attention_masks)
```

Then, we apply the tokenizer function to the train, validation and test sets.

This code block defines a function tokenize() that encodes the text data using the BERT tokenizer provided by the Hugging Face Transformers library. It then calls this function to tokenize the training, validation, and test data by encoding the input texts as input IDs and generating attention masks. These tokenized inputs are returned as numpy arrays.





```python
train_input_ids, train_attention_masks = tokenize(X_train, MAX_LEN)
val_input_ids, val_attention_masks = tokenize(X_valid, MAX_LEN)
test_input_ids, test_attention_masks = tokenize(X_test, MAX_LEN)
```

```python
def create_model(bert_model, max_len=MAX_LEN):
    
    ##params###
    opt = tf.keras.optimizers.Adam(learning_rate=1e-5)
    loss = tf.keras.losses.CategoricalCrossentropy()
    accuracy = tf.keras.metrics.CategoricalAccuracy()


    input_ids = tf.keras.Input(shape=(max_len,),dtype='int32')
    
    attention_masks = tf.keras.Input(shape=(max_len,),dtype='int32')
    
    embeddings = bert_model([input_ids,attention_masks])[1]
    
    output = tf.keras.layers.Dense(3, activation="softmax")(embeddings)
    
    model = tf.keras.models.Model(inputs = [input_ids,attention_masks], outputs = output)
    
    model.compile(opt, loss=loss, metrics=accuracy)
    
    
    return model
```

The code block creates a model using the pre-trained BERT model and the create_model() function defined earlier. It then prints the summary of the model's architecture.





```python
model = create_model(bert_model, MAX_LEN)
model.summary()
```
